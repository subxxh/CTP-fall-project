```python
import pandas as pd

fma_features_path = r"C:/Users/ross9/Downloads/fma_metadata/fma_metadata/features.csv" #Set this as the path to fma features file on your machine as it exceeds github file size limit
df_features = pd.read_csv(fma_features_path, low_memory=False, header = None)
df_features.head()
```

```python
heading_1 = df_features.iloc[0]
heading_2 = df_features.iloc[1]
heading_3 = df_features.iloc[2]

combined_cols = (heading_1.astype(str) + "_" + heading_2.astype(str) + "_" + heading_3.astype(str)).str.strip("_")

df_features = df_features.iloc[3:].copy()
df_features.columns = combined_cols

df_features = df_features.rename(columns={'track_id_nan_nan': 'track_id'})

df_features.head()
```

```python
df_features = df_features.astype("float32")
df_features['track_id'] = df_features['track_id'].astype("int32")
pd.set_option('display.float_format', '{:.10f}'.format)
df_features.head()
```

```python
for col in df_features.columns:
    print(col)
```

```python
relevant_features = [
    *[f"mfcc_mean_{i}" for i in range(1, 21)],
    *[f"mfcc_std_{i}" for i in range(1, 21)],

    *[f"chroma_cqt_mean_{i}" for i in range(1, 13)],
    *[f"chroma_cqt_std_{i}" for i in range(1, 12+1)],

    "spectral_centroid_mean_1",
    "spectral_centroid_std_1",
    "spectral_bandwidth_mean_1",
    "spectral_bandwidth_std_1",
    "spectral_rolloff_mean_1",
    "spectral_rolloff_std_1",

    "rmse_mean_1",
    "rmse_std_1",
    "zcr_mean_1",
    "zcr_std_1",
]
df_features = df_features[["track_id"] + relevant_features]
df_features.head()
```

```python
df_features.to_csv('./data/cleaned_fma_features.csv', index=False)
```
