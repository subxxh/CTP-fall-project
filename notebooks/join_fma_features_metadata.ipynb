```python
import pandas as pd

df_metadata = pd.read_csv('./data/cleaned_fma_metadata.csv')
df_features = pd.read_csv('./data/cleaned_fma_features.csv')
df_metadata.head()
```

```python
df_features.head()
```

```python
df_metadata["track_id"] = df_metadata["track_id"].astype(int)
df_features["track_id"] = df_features["track_id"].astype(int)

df_joined = df_metadata.merge(df_features, on="track_id", how="inner")
df_joined.head()
```

```python
df_joined.info(memory_usage='deep')
```

```python
df_joined.to_csv('./data/processsed/fma_metadata_features_joined.csv', index=False)
```
